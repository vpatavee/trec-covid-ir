```python
path_to_data = "../CORD-19/2020-07-16/"
path_to_metadata = "../CORD-19/2020-07-16/metadata.csv"
path_to_document_parses = "../CORD-17/2020-07-16/document_parses/"
```

```python
path_anserini = "anserini.covid-r5.abstract.qdel.bm25+rm3Rf.txt"
```

```python

with open(path_anserini, "r") as f:
    anserini_uid = f.readlines()
```

```python
anserini_uid[3].split()
```

```python


anserini_uid_set = set(e.split()[2] for e in anserini_uid if int(e.split()[3]) < 2000 )
```

```python
len(anserini_uid_set)
```

```python
abstract_list = list()
with open(path_to_metadata) as f_in:
    reader = csv.DictReader(f_in)
    for row in reader:
        cord_uid = row['cord_uid']
        if cord_uid not in anserini_uid_set:
            continue
        abstract = row['abstract']
        abstract_list.append({"uid": cord_uid, "text": abstract})
```

```python
abstract_list[0]
```

```python
for e in abstract_list:
    text = e["text"].split()
    text_list = list()
    i = 0
    while i + 50 < len(text):
        text_list.append(" ".join(text[i:i+200]))
        i += 150

    e["text_split"] = text_list
    
    
        
```

```python
abstract_list[0]
```

```python
m = list()
lines = list()
for e in abstract_list:
    uid = e["uid"]
    for t in e["text_split"]:
        m.append(uid)
        lines.append(t)
        
        




```
